```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv(r"/content/WA_Fn-UseC_-Telco-Customer-Churn.csv")


print(df.head())
print(df.info())


numeric_features = ['tenure', 'MonthlyCharges', 'TotalCharges']
print(df[numeric_features].describe())


df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')


plt.figure(figsize=(10, 6))
sns.countplot(x='Churn', data=df)
plt.title('Distribution of Churn')
plt.show()


for col in ['Contract', 'InternetService', 'tenure', 'MonthlyCharges']:
    plt.figure(figsize=(10, 6))
    if df[col].dtype == 'object':
        sns.countplot(x=col, hue='Churn', data=df)
    else:
        sns.histplot(data=df, x=col, hue='Churn', multiple="stack")
    plt.title(f'{col} Distribution by Churn')
    plt.show()


print(df.isnull().sum())

plt.figure(figsize=(12, 10))
sns.heatmap(df[numeric_features].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numeric Features')
plt.show()


categorical_features = ['Contract', 'InternetService', 'PaymentMethod']
for feature in categorical_features:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=feature, hue='Churn', data=df)
    plt.title(f'{feature} vs Churn')
    plt.show()
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].mean())
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split


df['Churn'] = (df['Churn'] == 'Yes').astype(int)

# One-Hot Encoding   
categorical_columns = df.select_dtypes(include=['object']).columns.drop('customerID')
df = pd.get_dummies(df, columns=categorical_columns, drop_first=True)

#   
scaler = StandardScaler()
df[numeric_features] = scaler.fit_transform(df[numeric_features])

df = df.drop('customerID', axis=1)
```

```python
from sklearn.model_selection import train_test_split
import torch


X = df.drop('Churn', axis=1)
y = df['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

#     PyTorch
X_train_tensor = torch.FloatTensor(X_train.astype(np.float32).values)
y_train_tensor = torch.FloatTensor(y_train.values).unsqueeze(1)
X_test_tensor = torch.FloatTensor(X_test.astype(np.float32).values)
y_test_tensor = torch.FloatTensor(y_test.values).unsqueeze(1)


print(f"Shape of X_train_tensor: {X_train_tensor.shape}")
print(f"Shape of y_train_tensor: {y_train_tensor.shape}")
print(f"Shape of X_test_tensor: {X_test_tensor.shape}")
print(f"Shape of y_test_tensor: {y_test_tensor.shape}")
```

```python
import torch.nn as nn
import torch.optim as optim

class ChurnNet(nn.Module):
    def __init__(self, input_dim):
        super(ChurnNet, self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(input_dim, 32),
            nn.ReLU(),
            nn.BatchNorm1d(32),
            nn.Dropout(0.5),
            nn.Linear(32, 16),
            nn.ReLU(),
            nn.BatchNorm1d(16),
            nn.Dropout(0.3),
            nn.Linear(16, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.layers(x)

model = ChurnNet(X_train.shape[1])
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)

#  
num_epochs = 100
for epoch in range(num_epochs):
    model.train()
    optimizer.zero_grad()
    outputs = model(X_train_tensor)
    loss = criterion(outputs, y_train_tensor)
    loss.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns


#    
with torch.no_grad():
    y_pred_prob = model(X_test_tensor)
    y_pred = (y_pred_prob > 0.5).float()


y_pred_np = y_pred.numpy()
y_test_np = y_test_tensor.numpy()


print('Accuracy:', accuracy_score(y_test_np, y_pred_np))
print('Precision:', precision_score(y_test_np, y_pred_np))
print('Recall:', recall_score(y_test_np, y_pred_np))
print('F1 Score:', f1_score(y_test_np, y_pred_np, average='macro'))


cm = confusion_matrix(y_test_np, y_pred_np)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()


with torch.no_grad():
    y_train_pred_prob = model(X_train_tensor)
    y_train_pred = (y_train_pred_prob > 0.5).float()

y_train_pred_np = y_train_pred.numpy()
y_train_np = y_train_tensor.numpy()

train_f1 = f1_score(y_train_np, y_train_pred_np, average='macro')
test_f1 = f1_score(y_test_np, y_pred_np, average='macro')

print(f"F1 Score on Training Set: {train_f1}")
print(f"F1 Score on Test Set: {test_f1}")

#   F1 Score
if train_f1 > test_f1:
    print("   . F1 Score    ,   .")
elif train_f1 < test_f1:
    print("         . F1 Score    .")
else:
    print("F1 Scores       ,        .")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc


model.eval()
with torch.no_grad():
    val_outputs = model(X_test_tensor)
    fpr, tpr, thresholds = roc_curve(y_test_tensor.cpu(), val_outputs.cpu())
    roc_auc = auc(fpr, tpr)


plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
